```
from google.colab import drive 
drive.mount('/content/gdrive')
import pandas as pd 
df=pd.read_csv('gdrive/My Drive/time_series_folder/Data/airline_passengers.csv', index_col= 0, parse_dates = True)
```

```
df = df.dropna()
```

```
df.head()
```

```
df.index
```

```
df["6-month_SMA"] = df["Thousands of Passengers"].rolling(window = 6).mean()
```

```
df["12-month_SMA"] = df["Thousands of Passengers"].rolling(window = 12).mean()
```

```
df["12-EWMA"] = df["Thousands of Passengers"].ewm(span = 12).mean()
```

```
df[["Thousands of Passengers", "6-month_SMA", "12-month_SMA"]].plot(figsize = (12, 5));
```

```
df[["Thousands of Passengers", "12-EWMA", "12-month_SMA"]].plot(figsize = (12, 5));
```
