```
import numpy as np
```

```
# we can see the recession happened during the 2008 and 2009
df["realgdp"].plot(figsize = (12, 5))
```

```
# Hodrick prescott filter -> separates the time series component into trend and cyclical component
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```
gdp_cycle, gdp_trend = hpfilter(df["realgdp"], lamb= 1600)
```

```
gdp_cycle.plot(figsize= (12, 5))
```

```
gdp_trend.plot(figsize = (12, 5))
```

```
df["trend"] = gdp_trend
```

```
df[["trend", "realgdp"]].plot(figsize= (12, 5))
```

```
# where the recession occured
df[["trend", "realgdp"]]["2005-01-01":].plot(figsize = (12, 5))
```
